```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
# create 2 group A and B, 
# Labels are -1 AND 1
groupA = [[1, 2], [2, 4] ,[3,3],[1,5],[8,1],[4,1],[7,1]]
groupALabel = [-1,-1,-1,-1,-1,-1,-1]



groupB = [[7, 6], [8, 4] ,[8,3],[5,5],[4,5],[4,7],[6,4]]
groupBLabel = [1,1,1,1,1,1,1]
```

```python
Ax=[]
Ay=[]

for i in groupA:
  Ax.append(i[0])
  Ay.append(i[1])


Bx=[]
By=[]

for i in groupB:
  Bx.append(i[0])
  By.append(i[1])


fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111)

ax.scatter(Ax, Ay, s=60, c='b', marker="s", label='Group A')
ax.scatter(Bx,By, s=60, c='r', marker="o", label='Group B')
plt.legend(loc='upper right');
plt.show()
```

```python
features  = groupA + groupB 
labels = groupALabel + groupBLabel
```

```python
def perceptron_prep(features):
  # appending one more column for each future ,1. every xi = (xi ,1)
  for feature in features:
    feature.append(1)

  #creating weight vector 
  weights = np.ones(len(features[0]))
  return (weights, features)
#Activation method
def perceptron_activation(number):
  if number > 0:
    return 1
  else:
    return -1
# Weight Training
def perceptron_weightTrainig(W,features, labels):
  while True:
    counter = 0
    for i in range(len(features)):
      predict = perceptron_activation(np.dot(W.T, features[i]))
      if predict * labels[i] <= 0 :
        W = W + np.dot(labels[i], features[i])
        counter = counter + 1
    if counter == 0:
      return W


  
```

```python
Weights, Features = perceptron_prep(features)
W = perceptron_weightTrainig(Weights,Features, labels)
```
